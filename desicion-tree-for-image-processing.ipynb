```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Directory containing the images
dataset_dir = "5x5_digit_images"

# Get the list of image files
image_files = sorted(os.listdir(dataset_dir))

# Display the images
plt.figure(figsize=(10, 5))
for i, file_name in enumerate(image_files):
    img_path = os.path.join(dataset_dir, file_name)
    img = Image.open(img_path)
    
    # Plot each image
    plt.subplot(2, 5, i + 1)
    plt.imshow(img, cmap="gray")
    plt.title(file_name.split(".")[0])  # Display the digit as title
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Classification function
def classify_digit(pattern):
    if pattern[3][1] == 1:
        return "Even"
    else:
        if pattern[0][2] == 1:
            return "Odd"
        else:
            return "Even"

# Example classification
for digit, pattern in digit_patterns.items():
    classification = classify_digit(pattern)
    print(f"Digit {digit} is classified as {classification}")

print(f"Images saved in the directory '{dataset_dir}'")
```
